```python
raga_count = {}
for val in data:
  if val[1] in raga_count:
    raga_count[val[1]] = raga_count[val[1]] + 1
  else:
    raga_count[val[1]] = 1

```

```python
# Print the samples that you have
raga_count
```

```python
# PERFORM DATA AUGMENTATION:
# aim for 200 samples for each raga]
import librosa 
import numpy as np
for sample in data: # loop through samples
  if raga_count[sample[1]] < 25:
    # do a pitch modulation, add it to data, and then update raga_count
    adj_pitch = sample[0]
    bins_per_octave = 12 # number of swaras (notes) that are being counted (totally)
    pitch_pm = 2
    pitch_change =  pitch_pm * 2 * np.random.uniform()
    adj_pitch = librosa.effects.pitch_shift(adj_pitch.astype('float64'), 22050, n_steps=pitch_change, bins_per_octave=bins_per_octave)
    
    data.append((adj_pitch, sample[1]))
    raga_count[sample[1]] = raga_count[sample[1]] + 1
```
