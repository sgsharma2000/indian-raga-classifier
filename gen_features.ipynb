```python
import tensorflow as tf
import numpy as np
import os
import pickle
from sklearn.model_selection import train_test_split
import librosa

# unpack the pickle
```

```python
dataset_dir = './music_dataset/'
pickle_file_path = os.path.join(dataset_dir, 'data_waug.pickle')

# unpack the pickle file
pickle_in = open(pickle_file_path, "rb")
test_data = pickle.load(pickle_in)
```

```python
len(test_data)
```

```python
# create mel chroma pickle
spec_cent_dataset = []
for index, val in enumerate(test_data):
  x = (val[0]).astype(float)
  #print(x)
  spec_cent = librosa.feature.chroma_stft(y=x, sr=22050)
  spec_cent_dataset.append((spec_cent, val[1]))

  if index % 100 == 0:
    print(index)
```

```python
file_path = os.path.join(dataset_dir, 'mel_chroma_pic_waug.pickle')
pickle_out = open(file_path, 'wb')
pickle.dump(spec_cent_dataset, pickle_out)
pickle_out.close()
```
